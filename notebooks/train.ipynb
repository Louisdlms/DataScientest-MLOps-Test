```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import pickle
import pandas as pd
import numpy as np
```

```python
# Ici, on développe un modèle de prédiction de plantes en fonction de caractéristiques 
# très simple sans approfondir la partie ML pour se concentrer sur la mise en production


iris = load_iris(as_frame = True)

print(iris.feature_names)
print(iris.target_names)

X, y = iris.data, iris.target

print(iris.frame)
```

```python
# Entraîner le modèle
model = RandomForestClassifier()
model.fit(X, y)
```

```python
# Sauvegarder le modèle
with open('../api/model.pkl', 'wb') as f:
    pickle.dump(model, f)
```
